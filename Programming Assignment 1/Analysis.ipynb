```python
#Algorithm 3
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python

def f1(n):
    if n<=2:
        return 1
    else:
        return f1(n-1) + f1(n-2)

def f2(n):
    if n<=2:
        return 1
    else:
        p=1
        q=1
        for i in range(3,n+1):
            r = p+q
            p = q
            q = r
        return r
    
G = np.array([[0,1],[1,1]])
def Power_of_G(n):
    if n==1:
        return G;
    else:
        if n%2 == 0:
            H = Power_of_G(n/2);
            return np.dot(H,H)
        else:
            H = Power_of_G((n-1)/2)
            return np.dot(H, np.dot(H,G))
        
def f3(n):
    temp = Power_of_G(n)
    return temp[0][1]
```

```python
inputs = np.arange(5, 30, 1)
inputs2 = np.arange(1,10000,1)
```

```python
def timecalc(algorithm, inputs):
    times=[]
    for i in inputs:
        start = float(time.time())
        answer = algorithm(i)
        end = float(time.time())
        runtime = float(end - start)
        times.append(runtime)
    return times
```

```python
f1_times = timecalc(f1, inputs)
plt.plot(inputs, f1_times, 'r') # plotting t, a separately 
plt.title('Runtime of F1()')
plt.xlabel('Value of N or Input')
plt.ylabel('Running Time')
```

```python
f2_times = []
f3_times = []
def timecalc_avg(algorithm, f_times):
    for i in range(1,10000):
        runtime = 0
        for j in range(0,100):
            start = float(time.time())
            answer = algorithm(i)
            end = float(time.time())
            this_time = float(end - start)
            runtime+=(this_time)
        f_times.append(runtime/100)
    return f_times

f2_times = timecalc_avg(f2, f2_times)
f3_times = timecalc_avg(f3, f3_times)

new_range = np
        
plt.plot(inputs2, f2_times, inputs2, f3_times) # plotting t, a separately 
plt.title('Running time of F2 vs F3')
plt.gca().legend(('F2','F3'))
plt.xlabel('Value of N or Input')
plt.ylabel('Running Time')

```
